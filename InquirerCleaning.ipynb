```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel("inquirerbasic.xls")
df.head()
```

```python
posWords = df[df.Positiv.notnull() | df.Pstv.notnull()]["Entry"].tolist()
negWords = df[df.Negativ.notnull() | df.Ngtv.notnull()]["Entry"].tolist()
```

```python
def clean_word_list(l):
    cleaned = []
    for word in l:
        if not word:
            word = "FALSE"
        pound_idx = word.find("#")
        if pound_idx > 0:
            cleaned.append(word[:pound_idx])
        else:
            cleaned.append(word)
    cleaned = sorted(list(set(cleaned)))  # get rid of duplicates
    return cleaned
```

```python
print(clean_word_list(posWords))
```

```python
posWords = clean_word_list(posWords)
negWords = clean_word_list(negWords)
```

```python
pd.Series(negWords).apply(type).value_counts()
```

```python
df_cleaned = pd.DataFrame(posWords)
df_cleaned["isPositive"] = 1
df_cleaned.columns=["Word","isPositive"]
```

```python
# df_cleaned
df_neg = pd.DataFrame(negWords)
df_neg["isPositive"] = 0
df_neg.columns=["Word","isPositive"]

df_cleaned = pd.concat([df_cleaned,df_neg])
```

```python
df_cleaned = df_cleaned.sort_values("Word")
```

```python
df_cleaned.reset_index(inplace=True)
```

```python
df_cleaned.drop("index",axis=1,inplace=True)
```

```python
def write_tag(isPos):
    if isPos == 0:
        return "neg"
    else:
        return "pos"
df_cleaned["Sentiment"] = df_cleaned["isPositive"].apply(write_tag)
```
